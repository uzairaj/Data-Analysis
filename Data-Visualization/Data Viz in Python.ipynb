```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

width = 0.15       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, bottom=men_means,label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```

```python
df_meal = pd.read_csv('archivedata/meal_info.csv') 
df_meal.head()
```

```python
df_center = pd.read_csv('archivedata/fulfilment_center_info.csv') 
df_center.head()
```

```python
df_food = pd.read_csv('archivedata/train.csv') 
df_food.head()
```

```python
df = pd.merge(df_food,df_center,on='center_id') 
df = pd.merge(df,df_meal,on='meal_id')
#df.to_csv('temp.csv', index=False)
```

```python
df.columns
```

```python
table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)
table
```

```python
table.plot(kind='bar')
```

```python
plt.bar(table.index,table['num_orders'])

plt.xticks(rotation=80) 

#x-axis labels 
plt.xlabel('Food item') 

#y-axis labels 
plt.ylabel('Quantity sold') 

#plot title 
plt.title('Most popular food') 

#save plot 
#plt.savefig('matplotlib_plotting_6.png',dpi=300,bbox_inches='tight') 

#display 
plt.show()
```

```python
download_url = ("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")

df = pd.read_csv(download_url)
df.head()
```

```python
df.columns
```

```python
df['Rank']
```

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"])
```

```python
df.Median.plot(kind="hist")
```

```python
medians = df[df["Median"] > 60000].sort_values("Median")
medians
medians.plot(x="Major", y=["P25th", "Median", "P75th"], kind="bar")
```

```python
df.plot(x="Median", y="Unemployment_rate", kind="scatter")
```

```python
cat=df.groupby("Major_category")["Total"].sum().sort_values()
cat
```

```python
cat.plot(kind="barh", fontsize=10)
```
